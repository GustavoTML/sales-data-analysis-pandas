```python
import pandas as pd
```

## Coletando dados do csv

```python
df_vendas = pd.read_csv("dataframe_nivel_1.csv", sep=",")
```

## Criando nova coluna a partir de outra

```python
df_vendas["valor_total"] = df_vendas["quantidade"] * df_vendas["preco_unitario"]
```

```python
df_vendas.head()
```

## Qual cliente comprou mais? (por quantidade)

```python
# Qual cliente comprou mais (por quantidade)
df_quant_por_cliente = df_vendas.groupby("cliente")["quantidade"].sum().sort_values(ascending=False)
df_quant_por_cliente.head(10)

# print(df_quant_por_cliente.idxmax()) <-- retorna so o nome do primeiro
```

Vemos que *Carlos* foi quem compro mais produtos, um total de 41.

## Qual cidade comprou mais produtos? (por quantidade)

```python
df_quant_por_cidade = df_vendas.groupby("cidade")["quantidade"].sum().sort_values(ascending=False)
df_quant_por_cidade.head(10)
```

*Salvador* foi a cidade que mais comprou, foi um total de 43 unidades.

## Venda Total da loja

```python
df_valor_total = df_vendas["valor_total"].sum()
print(f"R${df_valor_total:,.2f}")
```
