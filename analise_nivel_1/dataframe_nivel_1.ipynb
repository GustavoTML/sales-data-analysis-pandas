```python
import pandas as pd
```

## Coletando dados do csv

```python
df_vendas = pd.read_csv("dataframe_nivel_1.csv", sep=",")
```

## Parte 1 - Estatística básica

### Informações basicas do DataFrame

```python
df_vendas.describe()
```

```python
df_vendas.info()
```

### Média, Mediana e Moda

```python
quant_media = df_vendas["quantidade"].mean()
print(f"Média da quantidade de compra: {quant_media}")
```

```python
preco_mediana = df_vendas["preco_unitario"].median()
print(f"Mediana do preço unitário: {preco_mediana}")
```

```python
prod_moda = df_vendas["produto"].mode()
print(f"moda dos produtos: {prod_moda[0]}")
```

## Parte 2 - Ordenação

### Ordenação pelo preco_unitario

```python
df_preco_sort = df_vendas.sort_values(by="preco_unitario", ascending=False)
df_preco_sort.head()
```

### Compras com maior quantidade

```python
df_maior_quant = df_vendas.sort_values(by="quantidade", ascending=False)
df_maior_quant.head(5)
```

## Parte 3 - Filtros

### Vendas na cidade de são paulo

```python
df_sao_paulo = df_vendas[df_vendas["cidade"] == "São Paulo"]
df_sao_paulo.head()
```

### Venda com quantidade maior que 3

```python
df_quant_maior_que_3 = df_vendas[df_vendas["quantidade"] > 3]
df_quant_maior_que_3.head(10)
```

### Compras de mouse ou teclado

```python
df_mouse_ou_teclado = df_vendas[df_vendas["produto"].isin(["Mouse", "Teclado"])]
df_mouse_ou_teclado.head(10)
```

## Parte 4 - Consultas simples

### quantas vendas tem no dataframe?

```python
print(f"\nTotal de quantidade de vendas: {df_vendas.shape[0]}")
```

### Quantos produtos exitem?

```python
df_produtos = df_vendas["produto"].nunique() # ou .unique para retorna um lista dos produtos
print(f"\nProdutos diponíveis: {df_produtos}")
```

### Quantas vendas foram feitas por cada cidade?

```python
df_vendas_cidade = df_vendas.groupby("cidade")["quantidade"].sum()
df_vendas_cidade.head(5)
```

## Extras

#### Criando a coluna valor_total a partir da multiplicação das colunas quantidade e preco_unitario 

```python
df_vendas["valor_total"] = df_vendas["quantidade"] * df_vendas["preco_unitario"]
```

```python
df_vendas.head()
```

#### Qual cliente comprou mais? (por quantidade)

```python
# Qual cliente comprou mais (por quantidade)
df_quant_por_cliente = df_vendas.groupby("cliente")["quantidade"].sum().sort_values(ascending=False)
df_quant_por_cliente.head(10)

# print(df_quant_por_cliente.idxmax()) <-- retorna so o nome do primeiro
```

Vemos que *Carlos* foi quem compro mais produtos, um total de 41.

#### Qual cidade comprou mais produtos? (por quantidade)

```python
df_quant_por_cidade = df_vendas.groupby("cidade")["quantidade"].sum().sort_values(ascending=False)
df_quant_por_cidade.head(10)
```

*Salvador* foi a cidade que mais comprou, foi um total de 43 unidades.

#### Venda Total da loja

```python
df_valor_total = df_vendas["valor_total"].sum()
print(f"R${df_valor_total:,.2f}")
```
